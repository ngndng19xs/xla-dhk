```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def tien_xu_ly(image):
    """
    Hm tin x l nh cho vic nhn din bin s xe.
    Bao gm cc bc:
    1. Chuyn sang nh xm (Grayscale).
    2. Lm m Gaussian (GaussianBlur)  gim nhiu.
    
    :param image: nh u vo (c th l BGR 3 knh hoc Xm 1 knh).
    :return: nh  qua x l (nh xm  lm m).
    """
    
    # Bc 1: Chuyn nh sang nh xm (Grayscale)
    # Kim tra xem nh c phi l nh mu (3 knh) khng
    if len(image.shape) == 3 and image.shape[2] == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        # Nu nh  l nh xm (1 knh) th gn trc tip
        # iu ny gip hm hot ng ngay c khi u vo  l nh xm
        # v trnh li khi c gng cvtColor mt nh  l xm.
        gray_image = image

    # Bc 2: Lm m nh bng b lc Gaussian  gim nhiu
    # (5, 5) l kch thc kernel. Kch thc ny phi l s l.
    # 0 l gi tr sigmaX ( lch chun theo trc X).
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    
    return blurred_image

# --- Phn v d cch s dng hm ---
if __name__ == "__main__":
    #  chy v d ny, bn cn c mt nh tn l "bien_so_xe.jpg"
    # trong cng th mc.
    
    try:
        # Thay "bien_so_xe.jpg" bng ng dn thc t n nh ca bn
        image = cv2.imread("grey.png")
        
        # Kim tra xem nh c c ti thnh cng khng
        if image is None:
            print(f"Li: Khng th ti nh 'bien_so_xe.jpg'.")
            print("Vui lng kim tra li ng dn v m bo tp nh tn ti.")
        else:
            # 1. Gi hm tin x l
            processed_image = tien_xu_ly(image)

            # 2. Hin th kt qu bng Matplotlib
            plt.figure(figsize=(12, 6))

            # Hin th nh gc
            plt.subplot(1, 2, 1)
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            plt.title("nh Gc")
            plt.axis('off')

            # Hin th nh  x l
            plt.subplot(1, 2, 2)
            plt.imshow(processed_image, cmap='gray')
            plt.title("nh  Tin X L (Xm + M Gaussian)")
            plt.axis('off')

            plt.suptitle("Kt Qu Tin X L")
            plt.tight_layout()
            plt.show()
            
            print(" hon thnh v d tin x l.")

    except ImportError:
        print("\n--- LI ---")
        print("Vui lng ci t cc th vin cn thit:")
        print("pip install opencv-python")
        print("pip install matplotlib")
    except Exception as e:
        print(f"\n xy ra li: {e}")
        print("Hy m bo bn c tp nh hoc cc th vin  c ci t ng.")
```

```python
def phat_hien_canh_canny(path_img, pairs=None, save_dir="/mnt/data"):
    import cv2, numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path

    if pairs is None:
        pairs = [(30,90),(50,150),(75,150),(100,200),(150,250)]

    # c & m bo xm + lm m
    img = cv2.imread(path_img, cv2.IMREAD_UNCHANGED)
    if img is None:
        raise FileNotFoundError(f"Khng tm thy nh: {path_img}")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) if (len(img.shape)==3 and img.shape[2]==3) else img
    blur = cv2.GaussianBlur(gray, (5,5), 0)

    # hin th nh m tham chiu
    plt.figure(figsize=(5,4)); plt.imshow(blur, cmap='gray'); plt.title(f"Blur: {Path(path_img).name}"); plt.axis('off'); plt.show()

    # chy cc cp ngng
    save_dir = Path(save_dir); save_dir.mkdir(parents=True, exist_ok=True)
    outs = []
    for t1, t2 in pairs:
        edges = cv2.Canny(blur, t1, t2, apertureSize=3, L2gradient=True)
        out = save_dir / f"{Path(path_img).stem}_edges_{t1}_{t2}.png"
        cv2.imwrite(str(out), edges)
        outs.append(str(out))
        plt.figure(figsize=(5,4)); plt.imshow(edges, cmap='gray'); plt.title(f"{Path(path_img).name} - Canny ({t1},{t2})"); plt.axis('off'); plt.show()
    print(" lu:", *outs, sep="\n - ")

# V d:
phat_hien_canh_canny("bien_so_xe.jpg")
phat_hien_canh_canny("grey.png")
```
