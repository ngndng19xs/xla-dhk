```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def tien_xu_ly(image):
    """
    Hàm tiền xử lý ảnh cho việc nhận diện biển số xe.
    Bao gồm các bước:
    1. Chuyển sang ảnh xám (Grayscale).
    2. Làm mờ Gaussian (GaussianBlur) để giảm nhiễu.
    
    :param image: Ảnh đầu vào (có thể là BGR 3 kênh hoặc Xám 1 kênh).
    :return: Ảnh đã qua xử lý (ảnh xám đã làm mờ).
    """
    
    # Bước 1: Chuyển ảnh sang ảnh xám (Grayscale)
    # Kiểm tra xem ảnh có phải là ảnh màu (3 kênh) không
    if len(image.shape) == 3 and image.shape[2] == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        # Nếu ảnh đã là ảnh xám (1 kênh) thì gán trực tiếp
        # Điều này giúp hàm hoạt động ngay cả khi đầu vào đã là ảnh xám
        # và tránh lỗi khi cố gắng cvtColor một ảnh đã là xám.
        gray_image = image

    # Bước 2: Làm mờ ảnh bằng bộ lọc Gaussian để giảm nhiễu
    # (5, 5) là kích thước kernel. Kích thước này phải là số lẻ.
    # 0 là giá trị sigmaX (độ lệch chuẩn theo trục X).
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    
    return blurred_image

# --- Phần ví dụ cách sử dụng hàm ---
if __name__ == "__main__":
    # Để chạy ví dụ này, bạn cần có một ảnh tên là "bien_so_xe.jpg"
    # trong cùng thư mục.
    
    try:
        # Thay "bien_so_xe.jpg" bằng đường dẫn thực tế đến ảnh của bạn
        image = cv2.imread("grey.png")
        
        # Kiểm tra xem ảnh có được tải thành công không
        if image is None:
            print(f"Lỗi: Không thể tải ảnh 'bien_so_xe.jpg'.")
            print("Vui lòng kiểm tra lại đường dẫn và đảm bảo tệp ảnh tồn tại.")
        else:
            # 1. Gọi hàm tiền xử lý
            processed_image = tien_xu_ly(image)

            # 2. Hiển thị kết quả bằng Matplotlib
            plt.figure(figsize=(12, 6))

            # Hiển thị ảnh gốc
            plt.subplot(1, 2, 1)
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            plt.title("Ảnh Gốc")
            plt.axis('off')

            # Hiển thị ảnh đã xử lý
            plt.subplot(1, 2, 2)
            plt.imshow(processed_image, cmap='gray')
            plt.title("Ảnh Đã Tiền Xử Lý (Xám + Mờ Gaussian)")
            plt.axis('off')

            plt.suptitle("Kết Quả Tiền Xử Lý")
            plt.tight_layout()
            plt.show()
            
            print("Đã hoàn thành ví dụ tiền xử lý.")

    except ImportError:
        print("\n--- LỖI ---")
        print("Vui lòng cài đặt các thư viện cần thiết:")
        print("pip install opencv-python")
        print("pip install matplotlib")
    except Exception as e:
        print(f"\nĐã xảy ra lỗi: {e}")
        print("Hãy đảm bảo bạn có tệp ảnh hoặc các thư viện đã được cài đặt đúng.")
```

```python
def phat_hien_canh_canny(path_img, pairs=None, save_dir="/mnt/data"):
    import cv2, numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path

    if pairs is None:
        pairs = [(30,90),(50,150),(75,150),(100,200),(150,250)]

    # đọc & đảm bảo xám + làm mờ
    img = cv2.imread(path_img, cv2.IMREAD_UNCHANGED)
    if img is None:
        raise FileNotFoundError(f"Không tìm thấy ảnh: {path_img}")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) if (len(img.shape)==3 and img.shape[2]==3) else img
    blur = cv2.GaussianBlur(gray, (5,5), 0)

    # hiển thị ảnh mờ tham chiếu
    plt.figure(figsize=(5,4)); plt.imshow(blur, cmap='gray'); plt.title(f"Blur: {Path(path_img).name}"); plt.axis('off'); plt.show()

    # chạy các cặp ngưỡng
    save_dir = Path(save_dir); save_dir.mkdir(parents=True, exist_ok=True)
    outs = []
    for t1, t2 in pairs:
        edges = cv2.Canny(blur, t1, t2, apertureSize=3, L2gradient=True)
        out = save_dir / f"{Path(path_img).stem}_edges_{t1}_{t2}.png"
        cv2.imwrite(str(out), edges)
        outs.append(str(out))
        plt.figure(figsize=(5,4)); plt.imshow(edges, cmap='gray'); plt.title(f"{Path(path_img).name} - Canny ({t1},{t2})"); plt.axis('off'); plt.show()
    print("Đã lưu:", *outs, sep="\n - ")

# Ví dụ:
phat_hien_canh_canny("bien_so_xe.jpg")
phat_hien_canh_canny("grey.png")
```
