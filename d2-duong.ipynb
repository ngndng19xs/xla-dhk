```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def tien_xu_ly(image):
    """
    Hàm tiền xử lý ảnh cho việc nhận diện biển số xe.
    Bao gồm các bước:
    1. Chuyển sang ảnh xám (Grayscale).
    2. Làm mờ Gaussian (GaussianBlur) để giảm nhiễu.
    
    :param image: Ảnh đầu vào (có thể là BGR 3 kênh hoặc Xám 1 kênh).
    :return: Ảnh đã qua xử lý (ảnh xám đã làm mờ).
    """
    
    # Bước 1: Chuyển ảnh sang ảnh xám (Grayscale)
    # Kiểm tra xem ảnh có phải là ảnh màu (3 kênh) không
    if len(image.shape) == 3 and image.shape[2] == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        # Nếu ảnh đã là ảnh xám (1 kênh) thì gán trực tiếp
        # Điều này giúp hàm hoạt động ngay cả khi đầu vào đã là ảnh xám
        # và tránh lỗi khi cố gắng cvtColor một ảnh đã là xám.
        gray_image = image

    # Bước 2: Làm mờ ảnh bằng bộ lọc Gaussian để giảm nhiễu
    # (5, 5) là kích thước kernel. Kích thước này phải là số lẻ.
    # 0 là giá trị sigmaX (độ lệch chuẩn theo trục X).
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    
    return blurred_image

# --- Phần ví dụ cách sử dụng hàm ---
if __name__ == "__main__":
    # Để chạy ví dụ này, bạn cần có một ảnh tên là "bien_so_xe.jpg"
    # trong cùng thư mục.
    
    try:
        # Thay "bien_so_xe.jpg" bằng đường dẫn thực tế đến ảnh của bạn
        image = cv2.imread("grey.png")
        
        # Kiểm tra xem ảnh có được tải thành công không
        if image is None:
            print(f"Lỗi: Không thể tải ảnh 'bien_so_xe.jpg'.")
            print("Vui lòng kiểm tra lại đường dẫn và đảm bảo tệp ảnh tồn tại.")
        else:
            # 1. Gọi hàm tiền xử lý
            processed_image = tien_xu_ly(image)

            # 2. Hiển thị kết quả bằng Matplotlib
            plt.figure(figsize=(12, 6))

            # Hiển thị ảnh gốc
            plt.subplot(1, 2, 1)
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            plt.title("Ảnh Gốc")
            plt.axis('off')

            # Hiển thị ảnh đã xử lý
            plt.subplot(1, 2, 2)
            plt.imshow(processed_image, cmap='gray')
            plt.title("Ảnh Đã Tiền Xử Lý (Xám + Mờ Gaussian)")
            plt.axis('off')

            plt.suptitle("Kết Quả Tiền Xử Lý")
            plt.tight_layout()
            plt.show()
            
            print("Đã hoàn thành ví dụ tiền xử lý.")

    except ImportError:
        print("\n--- LỖI ---")
        print("Vui lòng cài đặt các thư viện cần thiết:")
        print("pip install opencv-python")
        print("pip install matplotlib")
    except Exception as e:
        print(f"\nĐã xảy ra lỗi: {e}")
        print("Hãy đảm bảo bạn có tệp ảnh hoặc các thư viện đã được cài đặt đúng.")
```

```python
# %% [markdown]
# PHÁT HIỆN CẠNH CANNY & IN ẢNH TRỰC TIẾP

import os, cv2, numpy as np, matplotlib.pyplot as plt

def phat_hien_canh(blur_img: np.ndarray, t1: int, t2: int,
                   apertureSize: int = 3, L2gradient: bool = True) -> np.ndarray:
    """Trả về ảnh nhị phân các cạnh (0/255) dùng Canny."""
    if blur_img.ndim == 3:
        gray = cv2.cvtColor(blur_img, cv2.COLOR_BGR2GRAY)
    else:
        gray = blur_img
    return cv2.Canny(gray, t1, t2, apertureSize=apertureSize, L2gradient=L2gradient)

# ---- Nạp ảnh (ưu tiên ảnh đã làm mờ sẵn) ----
blur_path = "grey.png"     # ảnh xám đã blur sẵn (từ d2-duong.ipynb)
src_path  = "bien_so_xe.jpg"

if os.path.exists(blur_path):
    blur = cv2.imread(blur_path, cv2.IMREAD_GRAYSCALE)
    blur = cv2.GaussianBlur(blur, (5,5), 0)  # làm mịn nhẹ (ổn định Canny)
    show_origin = blur.copy()
    origin_is_gray = True
    origin_title = "Ảnh xám đã làm mờ (grey.png)"
elif os.path.exists(src_path):
    src  = cv2.imread(src_path)            # BGR
    gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    show_origin = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)  # để hiển thị đúng màu
    origin_is_gray = False
    origin_title = "Ảnh gốc (đã tự chuyển xám + làm mờ)"
else:
    raise FileNotFoundError("Không tìm thấy /mnt/data/grey.png hoặc /mnt/data/bien_so_xe.jpg")

# ---- In ảnh đầu vào ----
plt.figure(figsize=(6,4))
plt.title(origin_title)
plt.imshow(show_origin, cmap="gray" if origin_is_gray else None)
plt.axis("off")
plt.show()

# Nếu ảnh gốc là RGB, in thêm ảnh xám đã làm mờ dùng cho Canny
if not origin_is_gray:
    plt.figure(figsize=(6,4))
    plt.title("Ảnh xám đã làm mờ sẽ dùng cho Canny")
    plt.imshow(blur, cmap="gray")
    plt.axis("off")
    plt.show()

# ---- Thử nghiệm nhiều cặp ngưỡng ----
threshold_pairs = [(30,90), (50,150), (70,170), (100,200), (120,240)]
for t1, t2 in threshold_pairs:
    edges = phat_hien_canh(blur, t1, t2)
    plt.figure(figsize=(6,4))
    plt.title(f"Canny edges (t1={t1}, t2={t2})")
    plt.imshow(edges, cmap="gray")
    plt.axis("off")
    plt.show()

# ---- Gợi ý tự động: t2 ≈ 2 * t1 (t1 ước lượng từ Otsu/hist) ----
# Lấy t1 theo Otsu rồi co lại chút để nhạy cạnh: t1 = 0.5 * ngưỡng Otsu
_, otsu_thr = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
t1_auto = max(10, int(0.5 * _))   # _ chính là ngưỡng Otsu trả về
t2_auto = min(255, 2 * t1_auto)

edges_auto = phat_hien_canh(blur, t1_auto, t2_auto)
plt.figure(figsize=(6,4))
plt.title(f"Canny (auto gợi ý: t1={t1_auto}, t2={t2_auto})")
plt.imshow(edges_auto, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# -*- coding: utf-8 -*-
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# =========================
# 1) HÀM PHÁT HIỆN CẠNH
# =========================
def phat_hien_canh(img_blur: np.ndarray, threshold1: int, threshold2: int,
                   apertureSize: int = 3, L2gradient: bool = True) -> np.ndarray:
    """
    Áp dụng Canny trên ảnh xám đã làm mờ.
    - img_blur: ảnh xám đã làm mờ (ndarray HxW)
    - threshold1: ngưỡng dưới
    - threshold2: ngưỡng trên
    - apertureSize: kích thước Sobel (3/5/7)
    - L2gradient: dùng chuẩn L2 để tính độ lớn gradient (nhạy hơn)
    Trả về: ảnh nhị phân (0/255) chỉ chứa các cạnh.
    """
    edges = cv2.Canny(img_blur, threshold1, threshold2,
                      apertureSize=apertureSize, L2gradient=L2gradient)
    return edges

# =========================
# 2) HÀM THỬ NGHIỆM NHIỀU NGƯỠNG
# =========================
def thu_nghiem_canny(img_blur: np.ndarray,
                     threshold_pairs=[(30, 90), (50, 150), (75, 150), (100, 200), (150, 250)],
                     save_dir="./"):
    """
    Chạy Canny với nhiều cặp ngưỡng, hiển thị và lưu ảnh kết quả.
    """
    save_dir = Path(save_dir)
    save_dir.mkdir(parents=True, exist_ok=True)

    # Tham chiếu: hiển thị ảnh xám đã làm mờ
    plt.figure(figsize=(6,4))
    plt.imshow(img_blur, cmap='gray')
    plt.title("Ảnh xám đã làm mờ (GaussianBlur)")
    plt.axis('off')
    plt.show()

    saved = []
    for (t1, t2) in threshold_pairs:
        edges = phat_hien_canh(img_blur, t1, t2)
        # Lưu file
        out_path = save_dir / f"edges_canny_{t1}_{t2}.png"
        cv2.imwrite(str(out_path), edges)
        saved.append(str(out_path))

        # Hiển thị
        plt.figure(figsize=(6,4))
        plt.imshow(edges, cmap='gray')
        plt.title(f"Canny edges (threshold1={t1}, threshold2={t2})")
        plt.axis('off')
        plt.show()

    print("Đã lưu:")
    for p in saved:
        print(" -", p)

# =========================
# 3) VÍ DỤ SỬ DỤNG
# =========================
# Trường hợp A: bạn đã có biến `img_blur` (ảnh xám đã làm mờ) từ file d2-duong.ipynb
# --> gọi trực tiếp:
# thu_nghiem_canny(img_blur, save_dir="/mnt/data")

# Trường hợp B: bạn chỉ có ảnh gốc, ta tự chuyển xám + làm mờ:
src_path = "bien_so_xe.jpg"      # đổi thành đường dẫn ảnh của bạn
img_bgr = cv2.imread(src_path)
if img_bgr is None:
    raise FileNotFoundError(f"Không tìm thấy ảnh: {src_path}")

img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (5, 5), 0)

thu_nghiem_canny(
    img_blur,
    threshold_pairs=[(30, 90), (50, 150), (75, 150), (100, 200), (150, 250)],
    save_dir="/mnt/data"                    # thư mục lưu ảnh kết quả
)
```

```python
def xu_ly_anh_canny(path_img, pairs=None, save_dir="/mnt/data"):
    import cv2, numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path

    if pairs is None:
        pairs = [(30,90),(50,150),(75,150),(100,200),(150,250)]

    # đọc & đảm bảo xám + làm mờ
    img = cv2.imread(path_img, cv2.IMREAD_UNCHANGED)
    if img is None:
        raise FileNotFoundError(f"Không tìm thấy ảnh: {path_img}")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) if (len(img.shape)==3 and img.shape[2]==3) else img
    blur = cv2.GaussianBlur(gray, (5,5), 0)

    # hiển thị ảnh mờ tham chiếu
    plt.figure(figsize=(5,4)); plt.imshow(blur, cmap='gray'); plt.title(f"Blur: {Path(path_img).name}"); plt.axis('off'); plt.show()

    # chạy các cặp ngưỡng
    save_dir = Path(save_dir); save_dir.mkdir(parents=True, exist_ok=True)
    outs = []
    for t1, t2 in pairs:
        edges = cv2.Canny(blur, t1, t2, apertureSize=3, L2gradient=True)
        out = save_dir / f"{Path(path_img).stem}_edges_{t1}_{t2}.png"
        cv2.imwrite(str(out), edges)
        outs.append(str(out))
        plt.figure(figsize=(5,4)); plt.imshow(edges, cmap='gray'); plt.title(f"{Path(path_img).name} - Canny ({t1},{t2})"); plt.axis('off'); plt.show()
    print("Đã lưu:", *outs, sep="\n - ")

# Ví dụ:
xu_ly_anh_canny("bien_so_xe.jpg")
xu_ly_anh_canny("grey.png")
```
